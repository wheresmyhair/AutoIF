```python
import jsonlines
from tqdm import tqdm
```

```python
original = []
with jsonlines.open('/root/autodl-tmp/projs/AutoIF/make_inst_following_deepseek/instf_volc_deepseek_with_original_ans.jsonl', 'r') as reader:
    for obj in reader:
        original.append(obj)
```

```python
original[0]
```

```python
original_idx_mapping = {}
for i, obj in enumerate(original):
    original_idx_mapping[obj['custom_id']] = i
```

```python
output_deepseek = []
output_raw = []
```

```python
with jsonlines.open('/root/autodl-tmp/projs/AutoIF/make_inst_following_deepseek/results.jsonl', 'r') as reader:
    for obj in tqdm(reader, total=len(original)):
        input_data = original[original_idx_mapping[obj['custom_id']]]
        if obj['custom_id'] != input_data['custom_id']:
            raise ValueError(f"Custom ID mismatch: {obj['custom_id']} != {input_data['custom_id']}")
        if input_data['original_ans'][0] == "idxerr":
            print(f"Skipping due to idxerr: {input_data['custom_id']}")
            continue
        out_deepseek = {
            "messages": input_data['body']['messages']+[obj['response']['body']['choices'][0]['message']],
        }
        out_raw = {
            "messages": input_data['body']['messages']+input_data['original_ans'],
        }
        output_deepseek.append(out_deepseek)
        output_raw.append(out_raw)
```
